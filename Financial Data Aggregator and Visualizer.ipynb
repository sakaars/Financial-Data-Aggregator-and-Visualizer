**Fetch Real-time Stock Market Data Using APIs**

```python
import requests
import pandas as pd

# Alpha Vantage API key (sign up on their website to get it)
api_key = 'IGPXY3F67BTN1T34'

# Define the API URL for real-time stock data
symbol = 'AAPL'  # Example: Apple stock
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={api_key}'

# Send a request to the API
response = requests.get(url)
data = response.json()

# Convert the response data into a pandas DataFrame
df = pd.DataFrame.from_dict(data['Time Series (Daily)'], orient='index')
df = df.astype(float)
df.head()
```
